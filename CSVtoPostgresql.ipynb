```python
from sqlalchemy import create_engine # 
from sqlalchemy.ext.declarative import declarative_base # nos va permitir tener acceso a las funcionalidades de ORM(Object Relation and Mapping) nos permite trabajar con objetos de python en lugar de queries de SQL directamente
from sqlalchemy.orm import sessionmaker

engine = create_engine('postgresql://dazufecar87:P1C1@localhost/homicidios')# sqlite:///newspaper.db') # nuestro motor, sqlite viene dentro de nuestra libreria de python

session = sessionmaker(bind=engine) # le pasamos nuestro motor

Base = declarative_base() # clase base de la cual van a extender todos nuestros modelos
```

```python
from sqlalchemy import Column, String, Integer, DateTime

class Homicides(Base):
    __tablename__ = 'homicidios_2018'
    
    id = Column(Integer, primary_key=True)
    fecha = Column(DateTime)
    departamento = Column(String)
    municipio = Column(String)
    dia = Column(String)
    hora = Column(String)
    barrio = Column(String)
    zona = Column(String)
    clase_sitio = Column(String)
    arma_empleada = Column(String)
    movil_agresor = Column(String)
    movil_victima = Column(String)
    edad = Column(Integer)
    genero = Column(String)
    estado_civil = Column(String)
    pais_nacimiento = Column(String)
    clase_empleado = Column(String)
    escolaridad = Column(String)
    codigo_dane = Column(Integer)
    

    def __init__(self,
                 index,
                 fecha,
                 departamento,
                 municipio,
                 dia,
                 hora,
                 barrio,
                 zona,
                 clase_sitio,
                 arma_empleada,
                 movil_agresor,
                 movil_victima,
                 edad,
                 genero,
                 estado_civil,
                 pais_nacimiento,
                 clase_empleado,
                 escolaridad,
                 codigo_dane):
        self.id = index
        self.fecha = fecha
        self.departamento = departamento
        self.municipio = municipio
        self.dia = dia
        self.hora = hora
        self.barrio = barrio
        self.zona = zona
        self.clase_sitio = clase_sitio
        self.arma_empleada = arma_empleada
        self.movil_agresor = movil_agresor
        self.movil_victima = movil_victima
        self.edad = edad
        self.genero = genero
        self.estado_civil = estado_civil
        self.pais_nacimiento = pais_nacimiento
        self.clase_empleado = clase_empleado
        self.escolaridad = escolaridad
        self.codigo_dane = codigo_dane
                 
```

```python
import logging
logging.basicConfig(level=logging.INFO)

import pandas as pd

from sqlalchemy.exc import DBAPIError

logger = logging.getLogger(__name__)


def main(filename):
    Base.metadata.create_all(engine) #Permite generar schema en nuestra db, Generar codigo bolder plate para poder configurar SQL alchemy en las bses de datos SQL normalmente se usan el concepto de sesion y de schema para poder determinar como estructura nuestros datos
    Session = session() # inicializar nuestra session
    homicidios_2018 = pd.read_csv(f'./modified_dataset/{filename}') # leemos articulos con pandas

    print(len(homicidios_2018))
    for index, row in homicidios_2018.iterrows(): #metodo de pandas que se llama iterrows, permite generar un loop adentro de cada una de nuestras filas de nuestra dataFrame
        #print(len(row)
        logger.info(f'Loading homicidio uid {index} into DB')
        homicidio = Homicides(index,
                            row['FECHA'],
                            row['DEPARTAMENTO'],
                            row['MUNICIPIO'],
                            row['DIA'],
                            row['HORA'],
                            row['BARRIO'],
                            row['ZONA'],
                            row['CLASE SITIO'],
                            row['ARMA EMPLEADA'],
                            row['MOVIL AGRESOR'],
                            row['MOVIL VICTIMA'],
                            row['EDAD'],
                            row['GENERO'],
                            row['ESTADO CIVIL'],
                            row['PAIS NACIMIENTO'],
                            row['CLASE EMPLEADO'],
                            row['ESCOLARIDAD'],
                            row['CODIGO DANE'],
                           )
        Session.add(homicidio) # esto mete nuestro articulo a la DB
  
    #try:
    Session.commit()
    # except (DBAPIError) as e:
    #   logger.warning('Error loading article', exc_info=False)

    Session.close()


if __name__ == '__main__':

    main('homicidios_2018.csv')
```
